```python
word_number = {
    'zero': 0,
    'one' : 1,
    'two' : 2,
    'three' : 3,
    'four' : 4,
    'five' : 5,
    'six' : 6,
    'seven' : 7,
    'eight' : 8,
    'nine' : 9
}
```

```python
def find_number(first_letter):

    if first_letter == 't':
        if str[0:3] == 'two':
            return ([word_number[str[0:3]]])
        elif str[0:5] == 'three':
            return ([word_number[str[0:5]]])
        else:
            return ("Invalid Input")


    elif first_letter == 'f':
        if str[0:4] == 'four':
            return ([word_number[str[0:4]]])
        elif str[0:4] == 'five':
            return ([word_number[str[0:4]]])
        else:
            return ("Invalid Input")

    elif first_letter == 's':
        if str[0:3] == 'six':
            return ([word_number[str[0:3]]])
        elif str[0:5] == 'seven':
            return ([word_number[str[0:5]]])
        else:
            return ("Invalid Input")

    elif first_letter == 'o':
        return ([word_number[str[0:3]]])


    elif first_letter == 'n':
        return ([word_number[str[0:4]]])


    elif first_letter == 'e':
        return ([word_number[str[0:5]]])
    
    elif first_letter == 'z':
        return ([word_number[str[0:4]]])
    else:
        return ('Invalid Input')
```
